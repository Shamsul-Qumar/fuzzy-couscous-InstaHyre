```python
# grouping Industry based on JobID.	AgeOfCompany, employees_count columns

df.groupby('Industry').agg({'JobID':'count', 'AgeOfCompany':'median', 'employees_count':'median'}).sort_values(by='JobID', ascending=False)
```

```python
df.Involvement=df.Involvement.str.split('→').str[0]
```

```python
df.to_csv('dbtable')
```

```python
# using dictionary to get the unique Location counts
dcta={}
for i in df.Location.str.split(','):
    for j in i:
        j=j.strip()
        if j not in dcta:
            dcta[j]=1
        else:
            dcta[j]+=1
```

```python
dcta=pd.DataFrame.from_dict(dcta, orient='index',columns=['Cnt'])
```

```python
dcta=dcta.sort_values(by='Cnt', ascending=False)
#dcta.to_csv('Location_Data')
```

```python
dcta.reset_index().to_csv('Location_Data')
```

```python
# using dictionary to get the unquie skills counts
dctb={}
for i in df.Skills.str.split(','):
    for j in i:
        j=j.strip()
        if j not in dctb:
            dctb[j]=1
        else:
            dctb[j]+=1
```

```python
dctb=pd.DataFrame.from_dict(dctb, orient='index',columns=['Cnt'])
```

```python
dctb=dctb.sort_values(by='Cnt', ascending=False)
```

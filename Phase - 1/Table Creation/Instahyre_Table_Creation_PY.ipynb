```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Table_622.csv')
```

```python
df.head(4)
```

```python
#Renaming columns

df.rename(columns={'name':'Name', 'estab_year':'Founded', 'employees_count':'Employees', 'location':'Location', 'job_ids':'JobID', 'skills':'Skills', 'hr':'HR'}, inplace=True)
```

```python
#Extracting Year from Founded

df.Founded=df.Founded.str.replace('[^0-9]', '', regex=True).astype('int64')
```

```python
df.JobID=df.JobID.astype('int64')
```

```python
#Creating Job Table

job_table= df.copy()[['JobID', 'Designation', 'Industry','Location', 'Skills', 'Name']]
```

```python
job_table
```

```python
#Creating Companies Table

companies_table=df.copy()[['Name','Founded','Employees']]
```

```python
#Dropping Duplicates

companies_table.drop_duplicates(inplace=True)
```

```python
#Resetting Indexes and Sorting
companies_table=companies_table.sort_values(by='Name').reset_index()
companies_table=companies_table.iloc[:,1:]
companies_table
```

```python
#Creating Details Table

details_table=df.copy()[['Skills', 'Involvement', 'Exp', 'HR']]
```

```python
#Dropping Duplicates
details_table.drop_duplicates(inplace=True)
```

```python
#Resetting Indexes and Sorting
details_table=details_table.sort_values(by='Skills').reset_index()
details_table=details_table.iloc[:,1:]
details_table
```

```python
#Generating primary key for companies table

c= companies_table.Name
cid=[]
for i in range(len(c)):
    cid.append(i)
companies_table['CompanyID']=cid
```

```python
#Generating primary key for details table
d= details_table.Skills
did=[]
for i in range(len(d)):
    did.append(i)
details_table['DetailID']= did
```

```python
details_table.Involvement=details_table.Involvement.str.split('→').str[0]
```

```python
details_table.Involvement.str.split('→').str[0].value_counts()
```

```python
details_table
```

```python
#Creating dictionaries to map Primary Keys to other tables
```

```python
dct1= details_table.set_index('Skills')['DetailID']
dct2=companies_table.set_index('Name')['CompanyID']
```

```python
job_table['DetailID']=job_table.Skills.replace(dct1)
job_table['CompanyID']=job_table.Name.replace(dct2)
```

```python
job_table.drop('Name', axis=1, inplace=True)
```
